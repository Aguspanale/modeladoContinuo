```python
a = 0.5
def derivadas(X):
  x, y  = X
  dotx = -y + a*x*(x**2 + y**2)
  doty = x + a*y*(x**2 + y**2)

  return np.array([dotx, doty])




w_1, w_2 = (-0.1, 0.1)
Y, X = np.mgrid[w_1:w_2:100j,w_1:w_2:100j]
x,y = derivadas([X, Y])
plt.figure(figsize=(12, 10), dpi=80)

plt.streamplot(X, Y, x, y, density=[0.5,0.5], broken_streamlines = False)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as wg
from scipy import integrate

@wg.interact(mu=(-5.0, 10.0, 0.1), x0=(-5.0, 5.0, 0.001), y0=(-5.0, 5.0, 0.001), x1=(-5.0, 5.0, 0.001), y1=(-5.0, 5.0, 0.001))
def graph(mu, x0, y0,x1,y1):
    def derivadasA(X, t):
        x, y = X
        dotx = y*t
        doty = x
        return np.array([dotx, doty])

    Nt = 1000
    tmax = 3
    
    t = np.linspace(0, tmax, Nt)
    X0 = [x0, y0]
    res = integrate.odeint(derivadasA, X0, t)
    res = res.T
    A = res[0]
    B = res[1]
    plt.plot(A, B)

    t = np.linspace(0, tmax, Nt)
    X0 = [x1, y1]
    res = integrate.odeint(derivadasA, X0, t)
    res = res.T
    A = res[0]
    B = res[1]


    plt.plot(A, B)
    plt.show()
```
